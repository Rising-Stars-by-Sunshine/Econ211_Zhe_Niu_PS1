```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
train
```

```python
test = pd.read_csv('test.csv')
test
```

```python
# Basic data exploration
print(train.head())
print(train.describe())
print(train.info())
```

```python
# Check for missing values
print(train.isnull().sum())
```

```python
# Replace missing values for V21 to V28 and Amount with median
for column in ['V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']:
    train[column].fillna(train[column].median(), inplace=True)

# Replace missing value in 'Class' with mode
train['Class'].fillna(train['Class'].mode()[0], inplace=True)

# Verify if all missing values are handled
print(train.isnull().sum())
```

```python
# fraudulent transactions
num_frauds = train[train['Class'] == 1].shape[0]
print(f"Number of fraudulent transactions: {num_frauds}")
```

```python
numeric_columns = (list(train.loc[:, 'Time':'Amount']))

fig = plt.figure(figsize=(20, 50))
rows, cols = 10, 3
for idx, num in enumerate(numeric_columns[:30]):
    ax = fig.add_subplot(rows, cols, idx+1)
    ax.grid(alpha = 0.7, axis ="both")
    sns.kdeplot(x = num, fill = True,color ="#3386FF",linewidth=0.6, data = train, label = "Train")
    sns.kdeplot(x = num, fill = True,color ="#EFB000",linewidth=0.6, data = test, label = "Test")
    ax.set_xlabel(num)
    ax.legend()
fig.tight_layout()
fig.show()
```
